```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def logloss(y_true, y_pred):
    ep = 1e-15
    y_pred_new = [max(i, ep) for i in y_pred]
    y_pred_new = [min(i, 1-ep) for i in y_pred]
    y_pred_new = np.array(y_pred_new)
    return -np.mean(y_true * np.log(y_pred_new) + (1-y_true)*np.log(1-y_pred_new))
```

```python
def gradient_descent(f1, f2, y_true, epochs):
    w1 = 1
    w2 = 2
    b = 0
    rate = 0.5
    loss_thresh = 0.4627
    n = len(f1)
    for i in range(epochs):
        weigh = w1 * f1 + w2 * f2 + b
        y_pred = sigmoid(weigh)
        y_pred = y_pred.astype('int')
        loss = logloss(y_true, y_pred)
        w1d = (1/n) * np.dot(np.transpose(f1),(y_pred - y_true))
        w2d = (1/n) * np.dot(np.transpose(f2),(y_pred - y_true))
        b_d = np.mean(y_pred-y_true)

        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        b = b - rate * b_d

        print(f"epoch = {i}, w1 = {w1}, w2 = {w2}, bias = {b}, loss={loss}")

        if loss<=loss_thresh:
            break

    return w1, w2, b
```

```python
gradient_descent(x_train['age'], x_train['affordibility'], y_train, 1000)
```
